```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
features = ['petal length (cm)',
            'petal width (cm)',
            'sepal length (cm)',
            'sepal width (cm)']

x = df.loc[:, features].values
```

```python
y = data.target
```

```python
x = df.loc[:, features].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(x, columns=features).plot.scatter('petal length (cm)', 'petal width (cm)')

plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
kmeans = KMeans(n_clusters=3, random_state=1)

kmeans.fit(x)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
x = pd.DataFrame(x, columns=features)
```

```python
colormap = np.array(['r', 'g', 'b'])
plt.scatter(x['petal length (cm)'], x['petal width (cm)'], c=colormap[labels])
plt.scatter(centroids[:,0], centroids[:,1], s=300, marker='x', c='k')

plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.scatter(x['petal length (cm)'], x['petal width (cm)'], c=colormap[labels])
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('K-Means Clustering (k = 3)')

plt.subplot(1, 2, 2)
plt.scatter(x['petal length (cm)'], x['petal width (cm)'], c=colormap[y], s=40)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.title('Flower Species')
plt.tight_layout()
```
